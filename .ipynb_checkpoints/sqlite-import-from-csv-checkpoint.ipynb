```python
import pandas as pd
import sqlite3
import glob, os, fnmatch
```

```python
matches = []
for root, dirnames, filenames in os.walk('/Users/Alexandre/Workstation/kaggle/movies'):
    for filename in fnmatch.filter(filenames, '*.csv'):
        matches.append(os.path.join(root, filename))
print matches
```

```python
connection = sqlite3.connect("/Users/Alexandre/Workstation/databases/fivethirtyeight.db")
```

```python
for file in matches:
    try:
        if '2015_01_14' not in file and '2015_01_30' not in file:
            tableName = file.split('/')[-1][:-4].replace('-', '_')
            df = pd.DataFrame(pd.read_csv(file, encoding='latin-1'))
            df.to_sql(tableName,connection, flavor='sqlite', if_exists='replace')
            print tableName + ' ... done'
    except Exception as e:
        print tableName
        print e
```

```python
connection.commit()
```
